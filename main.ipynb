```python
n_df = df.shape[0]
print(n_df)

df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

n_df2 = df.shape[0]
print(n_df2)
```

```python
print('Missing values:',df.isnull().sum())

# Drop rows where key columns are missing
df.dropna(subset=['Order Date', 'Sales', 'Profit', 'Quantity'], inplace=True)

# Fill missing text columns with 'Unknown'
df['Product Name']  = df['Product Name'].fillna('Unknown')
df['Category'] = df['Category'].fillna('Unknown')
df['Region']   = df['Region'].fillna('Unknown')

print('\nMissing values after:')
print(df.isnull().sum())
```

```python
# Convert Order Date from string to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=False)

# Ensure numeric columns are correct type
df['Sales']    = pd.to_numeric(df['Sales'],    errors='coerce')
df['Profit']   = pd.to_numeric(df['Profit'],   errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')

# Clean text columns — strip spaces and fix casing
df['Product Name']  = df['Product Name'].str.strip().str.title()
df['Category'] = df['Category'].str.strip().str.title()
df['Region']   = df['Region'].str.strip().str.title()

print('Data types after fixing:')
print(df.dtypes)
```

```python
df['Year']         = df['Order Date'].dt.year
df['Month']        = df['Order Date'].dt.month
df['Month Name']   = df['Order Date'].dt.month_name()
df['Quarter']      = df['Order Date'].dt.quarter
df['Day of Week']  = df['Order Date'].dt.day_name()

print('New date columns added:')
df[['Order Date', 'Year', 'Month', 'Month Name', 'Quarter']].head(5)
```

```python
# Profit Margin %
df['Profit Margin %'] = ((df['Profit'] / df['Sales']) * 100).round(2)

# Revenue per unit
df['Revenue per Unit'] = (df['Sales'] / df['Quantity']).round(2)

# Profit flag — is this order profitable?
df['Is Profitable'] = df['Profit'] > 0

print('New derived columns added:')
df[['Sales', 'Profit', 'Quantity', 'Profit Margin %', 'Revenue per Unit', 'Is Profitable']].head()
```

```python
# Remove rows where Sales or Quantity are 0 or negative (bad data)
before = df.shape[0]

df = df[df['Sales'] > 0]
df = df[df['Quantity'] > 0]

# Remove extreme outliers using IQR method on Sales
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 3 * IQR
upper = Q3 + 3 * IQR

df = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]
df.reset_index(drop=True, inplace=True)

after = df.shape[0]
print(f'Rows removed as outliers: {before - after}')
print(f'Clean dataset size: {after} rows')
```

```python
print(df.shape)
print(f'Missing Values : {df.isnull().sum().sum()}')
print(f'Duplicates     : {df.duplicated().sum()}')
print(f'Date Range     : {df["Order Date"].min().date()} to {df["Order Date"].max().date()}')
print(f'Categories     : {df["Category"].unique().tolist()}')
print(f'Regions        : {df["Region"].unique().tolist()}')
print(f'Total Sales    : ${df["Sales"].sum():,.2f}')
print(f'Total Profit   : ${df["Profit"].sum():,.2f}')

df.head()
```

```python
df.to_csv('sales_cleaned.csv', index=False)
print('Cleaned data saved')
```
