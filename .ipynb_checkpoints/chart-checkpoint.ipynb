```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import os
```

```python
# ââ Load cleaned dataset âââââââââââââââââââââââââââââââ
df = pd.read_csv('sales_cleaned.csv', encoding='latin-1')
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Ensure date columns exist
df['Year']       = df['Order Date'].dt.year
df['Month']      = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.month_name()
df['Quarter']    = df['Order Date'].dt.quarter

MONTH_ORDER = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

print(f'Loaded {df.shape[0]:,} rows Ã {df.shape[1]} columns')
df.head(3)
```

```python
# ââ Aggregate ââââââââââââââââââââââââââââââââââââââââââ
monthly = (
    df.groupby('Month Name')['Sales']
      .sum()
      .reindex(MONTH_ORDER)
      .reset_index()
)
monthly.columns = ['Month', 'Sales']

peak_idx = monthly['Sales'].idxmax()
peak_val = monthly['Sales'].max()
low_val  = monthly['Sales'].min()

# ââ Plot âââââââââââââââââââââââââââââââââââââââââââââââ
fig, ax = plt.subplots(figsize=(13, 5))

# Bar â highlight peak month in a different colour
bar_colors = [RED if i == peak_idx else BLUE for i in range(len(monthly))]
bars = ax.bar(monthly['Month'], monthly['Sales'],
              color=bar_colors, width=0.6, alpha=0.85, zorder=2)

# Trend line on top
ax.plot(monthly['Month'], monthly['Sales'],
        color='#0C447C', marker='o', linewidth=2,
        markersize=6, zorder=3, label='Sales trend')

# Shaded area under line
ax.fill_between(range(len(monthly)), monthly['Sales'],
                alpha=0.08, color=BLUE)

# Annotate peak
ax.annotate(
    f"Peak\n${peak_val/1000:.1f}K",
    xy=(peak_idx, peak_val),
    xytext=(peak_idx + 1.2, peak_val * 0.95),
    arrowprops=dict(arrowstyle='->', color=RED, lw=1.5),
    fontsize=9, color=RED, fontweight='bold'
)

# Average line
avg = monthly['Sales'].mean()
ax.axhline(avg, color=GRAY, linestyle='--', linewidth=1.2, label=f'Avg ${avg/1000:.1f}K')

# Formatting
ax.set_title('Monthly Sales Trend', pad=14)
ax.set_xlabel('Month')
ax.set_ylabel('Total Sales')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v/1000:.0f}K'))
ax.set_xticks(range(len(monthly)))
ax.set_xticklabels(monthly['Month'], rotation=40, ha='right')
ax.legend(fontsize=9)

# Subtitle
fig.text(0.13, 0.91,
         f"Peak month highlighted in red   |   Range: ${low_val/1000:.1f}K â ${peak_val/1000:.1f}K",
         fontsize=9, color=GRAY)

plt.tight_layout()
# plt.savefig('../images/chart1_monthly_trend.png', dpi=150, bbox_inches='tight')
plt.show()
# print('Saved â images/chart1_monthly_trend.png')
```

```python
# ââ Aggregate ââââââââââââââââââââââââââââââââââââââââââ
top10 = (
    df.groupby('Product Name')
      .agg(Sales=('Sales','sum'), Profit=('Profit','sum'))
      .sort_values('Sales', ascending=False)
      .head(10)
      .reset_index()
)
# Trim long product names for display
top10['Label'] = top10['Product Name'].apply(
    lambda x: x[:42] + 'â¦' if len(x) > 42 else x
)
top10 = top10.sort_values('Sales', ascending=True)  # flip for horizontal bar

# ââ Plot âââââââââââââââââââââââââââââââââââââââââââââââ
fig, ax = plt.subplots(figsize=(12, 7))

# Gradient-like colouring â darker = higher rank
n = len(top10)
color_shades = [plt.cm.Blues(0.4 + 0.55 * i / (n - 1)) for i in range(n)]

bars = ax.barh(top10['Label'], top10['Sales'],
               color=color_shades, height=0.65)

# Value labels at end of each bar
for bar, profit in zip(bars, top10['Profit']):
    w = bar.get_width()
    profit_tag = f"  ${w/1000:.1f}K  (Profit: ${profit/1000:.1f}K)"
    ax.text(w + w * 0.005, bar.get_y() + bar.get_height() / 2,
            profit_tag, va='center', fontsize=8.5, color='#2C2C2A')

# Vertical average line
avg_sales = top10['Sales'].mean()
ax.axvline(avg_sales, color=AMBER, linestyle='--',
           linewidth=1.3, label=f'Avg (top 10): ${avg_sales/1000:.1f}K')

# Formatting
ax.set_title('Top 10 Products by Revenue')
ax.set_xlabel('Total Sales')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v/1000:.0f}K'))
ax.tick_params(axis='y', labelsize=9)
ax.set_xlim(0, top10['Sales'].max() * 1.25)
ax.legend(fontsize=9)

fig.text(0.13, 0.91,
         'Darker bar = higher ranked product   |   Profit shown at bar end',
         fontsize=9, color=GRAY)

plt.tight_layout()
# plt.savefig('../images/chart2_top10_products.png', dpi=150, bbox_inches='tight')
plt.show()
# print('Saved â images/chart2_top10_products.png')
```

```python
# ââ Aggregate ââââââââââââââââââââââââââââââââââââââââââ
region = (
    df.groupby('Region')
      .agg(Sales=('Sales','sum'), Profit=('Profit','sum'), Orders=('Sales','count'))
      .sort_values('Sales', ascending=False)
      .reset_index()
)
region['Margin %'] = (region['Profit'] / region['Sales'] * 100).round(1)

# ââ Plot â grouped bar (Sales + Profit side-by-side) ââ
fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))

# ââ Left: grouped bars âââââââââââââââââââââââââââââââââ
x      = np.arange(len(region))
width  = 0.38

b1 = axes[0].bar(x - width/2, region['Sales'],   width, label='Sales',  color=BLUE,  alpha=0.9)
b2 = axes[0].bar(x + width/2, region['Profit'],  width, label='Profit', color=GREEN, alpha=0.9)

# Value labels
for bar in b1:
    h = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2, h + h*0.01,
                 f'${h/1000:.0f}K', ha='center', va='bottom', fontsize=8)
for bar in b2:
    h = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2, h + h*0.01,
                 f'${h/1000:.0f}K', ha='center', va='bottom', fontsize=8)

axes[0].set_title('Sales & Profit by Region')
axes[0].set_ylabel('Amount ($)')
axes[0].set_xticks(x)
axes[0].set_xticklabels(region['Region'])
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v/1000:.0f}K'))
axes[0].legend()

# ââ Right: pie chart with margin % inside âââââââââââââ
wedge_colors = PALETTE[:len(region)]
wedges, texts, autotexts = axes[1].pie(
    region['Sales'],
    labels=None,
    autopct='%1.1f%%',
    colors=wedge_colors,
    startangle=120,
    pctdistance=0.75,
    wedgeprops=dict(linewidth=1.5, edgecolor='white')
)
for at in autotexts:
    at.set_fontsize(10)
    at.set_fontweight('bold')

# Legend with margin info
legend_labels = [f"{row.Region}  (margin {row['Margin %']}%)" for _, row in region.iterrows()]
axes[1].legend(wedges, legend_labels, loc='lower center',
               bbox_to_anchor=(0.5, -0.12), fontsize=9, ncol=2)
axes[1].set_title('Revenue Share by Region')

fig.text(0.13, 0.93,
         'Margin % = Profit Ã· Sales Ã 100   |   Shown in legend',
         fontsize=9, color=GRAY)

plt.tight_layout()
plt.savefig('../images/chart3_region_sales.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved â images/chart3_region_sales.png')
```

```python
# ââ Pivot table: rows = Year, columns = Month ââââââââââ
heatmap_data = (
    df.groupby(['Year', 'Month Name'])['Sales']
      .sum()
      .reset_index()
      .pivot(index='Year', columns='Month Name', values='Sales')
      .reindex(columns=MONTH_ORDER)
)

# Shorten month names to 3-letter abbreviations for display
heatmap_data.columns = [m[:3] for m in heatmap_data.columns]

# ââ Plot âââââââââââââââââââââââââââââââââââââââââââââââ
fig, ax = plt.subplots(figsize=(14, max(3.5, len(heatmap_data) * 1.1)))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='.0f',
    cmap='YlOrRd',
    linewidths=0.6,
    linecolor='white',
    ax=ax,
    cbar_kws={'label': 'Sales ($)', 'shrink': 0.8},
    annot_kws={'size': 9}
)

# Format annotation text to show K values
for text in ax.texts:
    val = float(text.get_text().replace(',', ''))
    text.set_text(f'${val/1000:.1f}K')
    text.set_fontsize(8.5)

ax.set_title('Monthly Sales Heatmap â Year Ã Month')
ax.set_xlabel('')
ax.set_ylabel('Year')
ax.tick_params(axis='x', rotation=0)
ax.tick_params(axis='y', rotation=0)

fig.text(0.13, 0.96,
         'Darker cell = higher sales   |   Values shown in $K',
         fontsize=9, color=GRAY)

plt.tight_layout()
plt.savefig('../images/chart4_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved â images/chart4_heatmap.png')
```

```python
# ââ Aggregate ââââââââââââââââââââââââââââââââââââââââââ
cat = (
    df.groupby('Category')
      .agg(Sales=('Sales','sum'), Profit=('Profit','sum'))
      .reset_index()
)
cat['Margin %'] = (cat['Profit'] / cat['Sales'] * 100).round(1)
cat = cat.sort_values('Profit', ascending=False)

sub = (
    df.groupby(['Category', 'Sub-Category'])
      .agg(Sales=('Sales','sum'), Profit=('Profit','sum'))
      .reset_index()
)
sub['Margin %'] = (sub['Profit'] / sub['Sales'] * 100).round(1)
sub = sub.sort_values('Profit', ascending=True)

# ââ Plot â 2 panels ââââââââââââââââââââââââââââââââââââ
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# ââ Left: Category profit bars âââââââââââââââââââââââââ
bar_colors_cat = [GREEN if p > 0 else RED for p in cat['Profit']]
bars = axes[0].bar(cat['Category'], cat['Profit'],
                   color=bar_colors_cat, width=0.45)

# Margin % label inside/above each bar
for bar, margin in zip(bars, cat['Margin %']):
    h  = bar.get_height()
    yp = h + (h * 0.02 if h >= 0 else h * 0.02)
    axes[0].text(bar.get_x() + bar.get_width()/2, yp,
                 f'{margin}%\nmargin',
                 ha='center', va='bottom', fontsize=9,
                 fontweight='bold',
                 color=GREEN if h > 0 else RED)

axes[0].axhline(0, color='black', linewidth=0.8)
axes[0].set_title('Total Profit by Category')
axes[0].set_ylabel('Total Profit ($)')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v/1000:.0f}K'))

# ââ Right: Sub-category horizontal bar ââââââââââââââââ
sub_colors = [GREEN if p > 0 else RED for p in sub['Profit']]
bars2 = axes[1].barh(sub['Sub-Category'], sub['Profit'],
                     color=sub_colors, height=0.6, alpha=0.88)

# Value labels
for bar in bars2:
    w  = bar.get_width()
    xp = w + abs(sub['Profit'].max()) * 0.01 if w >= 0 else w - abs(sub['Profit'].max()) * 0.01
    ha = 'left' if w >= 0 else 'right'
    axes[1].text(xp, bar.get_y() + bar.get_height()/2,
                 f'${w/1000:.1f}K', va='center', ha=ha, fontsize=8)

axes[1].axvline(0, color='black', linewidth=0.8)
axes[1].set_title('Profit by Sub-Category')
axes[1].set_xlabel('Total Profit ($)')
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v/1000:.0f}K'))
axes[1].tick_params(axis='y', labelsize=8.5)

# Colour legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=GREEN, label='Profitable'),
                   Patch(facecolor=RED,   label='Loss-making')]
axes[1].legend(handles=legend_elements, loc='lower right', fontsize=9)

fig.text(0.13, 0.93,
         'Green = profit   |   Red = loss   |   Margin % shown on category bars',
         fontsize=9, color=GRAY)

plt.suptitle('Category & Sub-Category Profitability', fontsize=15,
             fontweight='bold', y=1.0)
plt.tight_layout()
plt.savefig('../images/chart5_profit_category.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved â images/chart5_profit_category.png')
```
